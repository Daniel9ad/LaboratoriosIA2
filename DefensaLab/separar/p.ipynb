```python
import pandas as pd
import shutil
import requests
from PIL import Image
import io
```

```python
df2 = pd.read_csv('styles.csv', index_col=0)
df2Link = pd.read_csv('images2.csv')
```

```python
set(df2['subCategory'].tolist())
```

```python
dfTipo2 = df2.loc[df2['subCategory']=='Bottomwear'].copy()
set(dfTipo2['articleType'].tolist())
```

```python
dfTipo2 = df2.loc[df2['articleType']=='Belts'].copy()
dfTipo2
```

```python
for img in dfTipo2.index.tolist():
    archivo = f'{img}.jpg'
    link = df2Link.loc[df2Link['filename']==archivo]['link'].values[0]
    print(link)
    response = requests.get(link)
    # Verifica si la solicitud fue exitosa (cÃ³digo de estado 200)
    if response.status_code == 200:
        # Abre la imagen descargada desde la respuesta
        imagen = Image.open(io.BytesIO(response.content))
        # Redimensiona la imagen a un tamaÃ±o especÃ­fico (por ejemplo, 300x300 pÃ­xeles)
        nuevo_tamano = (400, 500)
        imagen_redimensionada = imagen.resize(nuevo_tamano)
        # Guarda la imagen redimensionada en el disco
        imagen_redimensionada.save(f'C:/Users/44792/Downloads/Clothing_dataset/Belt/{archivo}')
        print("Imagen descargada y redimensionada exitosamente")
    else:
        print("Error al descargar la imagen. CÃ³digo de estado:", response.status_code)
```
