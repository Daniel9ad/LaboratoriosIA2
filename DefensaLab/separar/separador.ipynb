```python
import pandas as pd
import shutil
import requests
from PIL import Image
import io
```

```python
df = pd.read_csv('images.csv')
df
```

```python
df2 = pd.read_csv('styles.csv', index_col=0)
```

```python
df2Link = pd.read_csv('images2.csv')
```

```python
set(df2['subCategory'].tolist())
```

```python
dfTipo2 = df2.loc[df2['articleType']=='Sports Shoes'].copy()
dfTipo2
```

```python
for img in dfTipo2.index.tolist()[:400]:
    archivo = f'{img}.jpg'
    link = df2Link.loc[df2Link['filename']==archivo]['link'].values[0]
    print(link)
    response = requests.get(link)
    # Verifica si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Abre la imagen descargada desde la respuesta
        imagen = Image.open(io.BytesIO(response.content))
        # Redimensiona la imagen a un tamaño específico (por ejemplo, 300x300 píxeles)
        nuevo_tamano = (400, 500)
        imagen_redimensionada = imagen.resize(nuevo_tamano)
        # Guarda la imagen redimensionada en el disco
        imagen_redimensionada.save(f'C:/Users/44792/Downloads/Clothing_dataset/Shoes/{archivo}')
        print("Imagen descargada y redimensionada exitosamente")
    else:
        print("Error al descargar la imagen. Código de estado:", response.status_code)
```
